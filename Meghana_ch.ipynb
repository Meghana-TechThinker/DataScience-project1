```python
import matplotlib.pyplot as plt
import numpy as np
exam_scores = np.random.randint(0,100,150)
plt.hist(exam_scores, bins=20, cumulative=True, color='purple', edgecolor='black')
plt.xlabel('Exam Scores')
plt.ylabel('number of Students')
plt.title('Cumulative Histogram of Exam Scores')
plt.show()
```

```python
#stacked histogram with Multiple datasets
import matplotlib.pyplot as plt
import numpy as np
marks1=np.array([2,4,5,79,10,11,13,14,15])
marks2=np.array([6,7,8,10,11,12,13,14,15,16])
plt.hist([marks1,marks2],bins=10, stacked=True, color=['orange','skyblue'],edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.title('Stacked histogram with multiple datasets')
plt.legend(['Mark dataset1','Mark dataset2'])
plt.show()
```

```python
import pandas as pd
#integer series
int_series=pd.Series([1,2,3,4],index=["a","b","c","d"])
print(int_series)
print('-----------------')
#float series
float_series=pd.Series([1.1,2.2,3.3,4.4])
print(float_series)
print('-----------------')
#String series
string_series=pd.Series(['a','b','c','d'])
print(string_series)
print('-----------------')
#label based indexing
data={'a':10,'b':20,'c':30}
series=pd.Series(data)
print(series)
print(series['a'])
print(series[['a','c']])
```

```python
data={'Name':['Alice','bob','Ave'],
      'Age':[25,30,35],
      'city':['New york','Los Angeles','Chicago']}
df=pd.DataFrame(data)
print(df)
print('------------------')
#selecting columns')
print(df.Name)
print(df['Name'])
print('-------------------')
print(df[['Name','city']])
print('----------------')
#select a single row by label
print("1.\n",df.loc[0])
print("2.\n",df.iloc[0])
print('--------------')
#conditional selection
print('conditional')
print(df[df['Age']>25])
print(df.loc[(df.Name=='bob')&(df.city=='Los Angeles')])
```

```python
import pandas as pd
data={
    'name':['Alice','Bob','charlie','Alice','Bob'],
    'age':[10,20,30,10,20],
    'city':['Newyork','francisco','los angeles','Newyork','francisco'],
    'salary':[70000,80000,120000,70000,80000]
}
df=pd.DataFrame(data)
grouped=df.groupby('name')
print(grouped.groups)
print("**************")
#calculate the sum of salaries for each group
salary_sum= grouped['salary'].sum()
print(salary_sum)
print("*****************")
age_mean=grouped['age'].mean()
print(age_mean)
print("------------------")
#Grouping by multiple columns
#group by 'name' and 'city'
grouped_multiple=df.groupby(['name','city'])
#calculate the sum of salaries for each group
salary_sum_multiple=grouped_multiple['salary'].sum()
print(salary_sum_multiple)
```

```python
#pandas merging,joining,concatenation of the DataFrame
import pandas as pd
data1={'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32],
       'Address':['Nagpur','kanpur','Allahabad','karnata'],
       'Qualification':['Msc','MA','MCA','Phd']}
data2={'Name':['Abhi','Sanju','Gautham','Anuhya'],
       'Age':[17,14,26,36],
       'Address':['Nagpur','kanpur','Allahabad','karnata'],
       'Qualification':['B.Tech','M.Tech','BA','Bcom']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df)
print(df1)
print(df, "\n\n", df1)
res1=pd.concat([df,df1])
res1
res2=pd.concat([df,df1],axis=1, join='inner')
res2
```

```python
#concatenating DataFrame by ignoring indexes:
import pandas as pd
data1={'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32],
       'Address':['Nagpur','kanpur','Allahabad','karnata'],
       'Qualification':['Msc','MA','MCA','Phd']}
data2={'Name':['Abhi','Sanju','Gautham','Anuhya'],
       'Age':[17,14,26,36],
       'Address':['Nagpur','kanpur','Allahabad','karnata'],
       'Qualification':['B.Tech','M.Tech','BA','Bcom']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df, "\n\n", df1)
res=pd.concat([df,df1], ignore_index=True)
res

```

```python
#
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32]}
data2={'key':['k0','k1','k2','k3'],
       'Name':['Abhi','Sanju','Gautham','Anuhya'],
       'Address':['Nagpur','kanpur','Allahabad','karnata'],
       'Qualification':['B.Tech','M.Tech','BA','Bcom']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df, "\n\n", df1)
res=pd.merge(df, df1, on='key')
res
```
